```python
from pathlib import Path
```

```python
dinet_trained_videos = []

with open("../training_video_name.txt", "r") as f:
    for line in f:
        dinet_trained_videos.append(line.strip())
```

```python
dinet_trained_videos
```

```python
dataset_videos = list(Path("../datasets/").glob("*.mp4"))
```

```python
in_dinet = []
not_in_dinet = []
for video in dataset_videos:
    clip_num = str(int(video.stem.split("_")[-1]))
    video_name = "_".join(video.stem.split("_")[1:-1]+[clip_num+".mp4"])
    if video_name in dinet_trained_videos:
        in_dinet.append(video)
    else:
        not_in_dinet.append(video)
```
